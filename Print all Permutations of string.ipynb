```python
# we print only when i == length of string
# j is for looping

def permute(data, i, length): 
    if i==length: 
        print(''.join(data) )
    else: 
        for j in range(i,length):                     # j is for looping 
            data[i], data[j] = data[j], data[i] 
            permute(data, i+1, length)                # i is for data
            data[i], data[j] = data[j], data[i]  


string = "ABCD"
n = len(string) 
data = list(string) 
permute(data, 0, n)
```
